$f(x) = x^2 e^{-x^2}$

```python
# define function f

x = sp.Symbol('x',real=True)
f = x**2*sp.exp(-x**2)
print('Following functions is:',f)

# differentiate f

f_diff = f.diff()
f_diff2 = f_diff.diff()
print("Derivative of ' is:", f_diff)
print("Derivative of '' is:",f_diff2)

# solve f' = 0

sol = sp.solve(sp.Eq(f_diff,0))
print("Solutions points are:", sol)

# stating the conclusions

print('Because this solution is < 0, we have a Max for -1:', f_diff2.subs(x,-1))
print('Because this solution is > 0, we have a Min for 0:',f_diff2.subs(x,0))
print('Because this solution is < 0, we have a Max for 1:',f_diff2.subs(x,1))

# za da vidime kolku se tocno tie vrednosti, mora kriticnite tocki sto gi dobivme, 
# da gi zamenime so ogrinalnata funkcija

print('For -1, we have Max value:', f.subs(x,-1))
print('For 0, we have Min value:',f.subs(x,0))
print('For 1, we have Max value:',f.subs(x,1))

# graph the function f
sp.plot(f, (x, -2.5, 2.5), ylabel='y')
```

## HW Problem 1

The following diagram shows a rectangular area ABCD enclosed on three sides by 60 m of fencing, and on the fourth by a wall AB.

image.jpeg


Find the dimensions of the rectangle that enclose maximum area, and then find the maximum area.

```python
# Define variables and the two functions: perimeter and area function

a = sp.Symbol('a',real=True)
b = sp.Symbol('b',real=True)

# MAXIMIZE(Area)=a*b
# PERIMETAR => 2*a+b=60
# f=a*b => a+b+a=60 => 2a+b=60 => 2a=60-b => a=60-b/2 => a=60/2-b/2 => a=30-b/2
# f=a*b => (30-b/2)*(b/1) => 30*b-1/2*b**2
# f=30*b-1/2*b**2

f=30*b-1/2*b**2
print('Following functions is:',f)

# differentiate f

f_diff = f.diff()
f_diff2 = f_diff.diff()
print("Derivative of ' is:", f_diff)
print("Derivative of '' is:",f_diff2)

# solve f' = 0

sol = sp.solve(sp.Eq(f_diff,0))
print("Side b is:", sol)

# stating the conclusions

# 2*a+b=60
# 2*a+30=60 => 2*a=60-30 => a=60-30/2 => a=30/2 => a=15

print('Because this solution is < 0, we have a Max for side b = 30:', f_diff2.subs(15,30))
print('Max area is:', f.subs(b,30))
```

The time taken, $T$, measured in tenths of a second, for the crocodile to reach its prey is given by
\begin{equation}T(x)=7 \sqrt{36+x^2} + 4(20-x)\end{equation}

* Calculate the time taken if the crocodile does not travel on land.
* Calculate the time taken if the crocodile swims the shortest distance possible.
* Between these two extremes there is one value of $x$ which minimizes the time taken. Find this value of $x$ and the minimum possible time. Use Gradient Descent Method with appropriate choice of $x_0$ and $\alpha$.

```python
# Define variables and functions

x = sp.Symbol('x',real=True)
t=7*sp.sqrt(36+x**2)+4*(20-x)
print('Following functions is:',t)

t_diff = t.diff()
t_diff2 = t_diff.diff()
print("Derivative of ' is:", t_diff)
print("Derivative of '' is:",t_diff2)

# Lambdify the functions

t = sp.lambdify(x,t) # f-jata ja pretvara od Sympy vo Python
t_diff = sp.lambdify(x,t_diff) # prv izvod so lambdify
t_diff2 = sp.lambdify(x,t_diff2) # vtor izvod so lambdify

# Calculate the time taken if the crocodile does not travel on land.

print('Time taken when crocodile does not travel on land:', t(20))
      
# Calculate the time taken if the crocodile swims the shortest distance possible.

print('Time taken when crocodile swims the shortest distance possible.:', t(0))
      
# # Finding the minimum time

alpha = 0.5
xk = 5
tol = 1e-7
max_iter = 30
k = 0

# Run gradient descent calculations

print("k\t xk\t\t t(xk)\t\t t'(xk)")
print('-------------------------------------------------------')

for k in range(max_iter):
    print(f'{k}\t {np.round(xk, 4)}\t\t {np.round(t(xk), 4)}\t\t {np.round(t_diff(xk), 4)}')
    xk = xk - alpha*t_diff(xk)
```
